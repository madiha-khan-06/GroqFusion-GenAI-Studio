# 🤖 Introduction to Generative AI, Agentic AI, LLMs, APIs & Frameworks

---

## 🌟 What is Generative AI?

**Generative AI** is a type of Artificial Intelligence that can create **new content** such as text, images, videos, audio, or code by learning patterns from large datasets.

👉 **Simple Definition:**  
Generative AI is AI that **generates new content based on learned data**.

---

### 🔑 Key Uses

- ✍️ Content Writing (blogs, emails, reports)  
- 💬 Chatbots & Virtual Assistants  
- 🎨 Image Generation  
- 💻 Code Generation  

---

## 🤖 What is Agentic AI?

**Agentic AI** refers to AI systems that can **act independently**, make decisions, and perform tasks step-by-step to achieve a specific goal.

👉 **Simple Definition:**  
Agentic AI is AI that can **think, plan, and take actions on its own** to complete tasks.

---

### 🔑 Key Features

- 🧠 Decision-Making Ability  
- 🎯 Goal-Oriented Behavior  
- 🔄 Multi-Step Task Execution  
- 🔗 Uses Tools (APIs, databases)  
- 🤖 Minimal Human Intervention  

---

## 🔗 What is LangChain?

**LangChain** is a framework used to build applications powered by **Large Language Models (LLMs)** like GPT.

👉 **Simple Definition:**  
LangChain helps developers **connect LLMs with data, tools, and workflows** to build intelligent AI applications.

---

## 🏗️ What is a Framework?

A **framework** is a pre-built structure that helps developers build applications quickly and efficiently.

It provides:
- 🧰 Tools  
- 📏 Rules  
- ♻️ Reusable Components  

👉 **Simple Definition:**  
A framework is a **ready-made structure** that helps developers build applications **faster and in an organized way**.

---

### 📌 Examples

- 🌐 Django (Web Development)  
- ⚛️ React (Frontend Development)  

---

## 🧠 What is an LLM (Large Language Model)?

An **LLM (Large Language Model)** is an advanced AI model trained on massive amounts of text data to understand and generate human-like language.

👉 **Simple Definition:**  
LLM is an AI trained on huge text data to **understand and generate human language**.

---

### 🔑 Key Features of LLM

- 🧠 Natural Language Understanding (NLU)  
- ✍️ Human-like Text Generation  
- 🔄 Context Awareness  
- 🌍 Multilingual Support  
- ⚡ Scalable & Adaptable  

---

### 🚀 What LLM Can Do

- 💬 Chatbots & Assistants  
- ✍️ Content Writing  
- 💻 Code Generation  
- 🌐 Language Translation  
- 📊 Data Summarization  

---

## 🔌 What is an API?

**API (Application Programming Interface)** is a set of rules that allows different applications to **communicate and exchange data**.

👉 **Simple Definition:**  
API is a bridge that helps **two systems talk to each other**.

---

### 🌍 Real-World Uses

- 🌦️ Weather Apps  
- 💳 Payment Gateways  
- 🔐 Login Authentication  
- 🗺️ Maps & Navigation  
- 🛒 E-commerce Platforms  
- 🤖 AI Applications  

---

### 🔑 Types of APIs

- 🌐 Web APIs (REST, SOAP)  
- 🔐 Private APIs (internal use)  
- 🤝 Public APIs (open for developers)  
- 🔗 Partner APIs (shared with partners)  

---

## 🔄 API Use Case (Real-Life Example)

### 🍔 Food Delivery App Flow

1️⃣ User selects food  
2️⃣ App calls Restaurant API  
3️⃣ App calls Payment API  
4️⃣ Uses Maps API for tracking  
5️⃣ Displays live order status  

---

## 🔑 What is an API Key?

An **API Key** is a unique code used to **authenticate and authorize** access to an API.

👉 **Simple Definition:**  
API Key is like a **password** that allows you to use an API securely.

---

## 🎯 Final Takeaway

- 🤖 Generative AI → Creates content  
- 🧠 Agentic AI → Thinks & acts independently  
- 🔗 LangChain → Connects LLMs with tools  
- 🏗️ Framework → Pre-built structure for development  
- 🧠 LLM → Understands & generates language  
- 🔌 API → Connects systems  
- 🔑 API Key → Secures API access

```python
import os
import getpass
from langchain_groq import ChatGroq
```

```python
# 1. Set your Groq API Key (Get it at https://console.groq.com)
os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
# 2. Initialize the Groq model (Llama 3.3 is currently a top free choice)
groq_chat = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
# 3. Simple Test
response = groq_chat.invoke("Explain Quantum Computing in one sentence.")
print(f"Groq Response: {response.content}")
print("✅ Groq connector is ready!")
```

```python
response = groq_chat.invoke("what is APIs and Its use case")
print(response.content)
```
